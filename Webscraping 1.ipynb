# 1. Start by performing a GET request on the url above and convert the response into a BeautifulSoup object.

```python
url = 'https://realpython.github.io/fake-jobs/'

response = requests.get(url)
```

```python
type(response)
```

```python
soup = BS(response.text)

soup
```

---

1a. Use the .find method to find the tag containing the first job title ("Senior Python Developer").

Extract the text from this title.

```python
soup.find('h2').text
```

---

1b. Now, use what you did for the first title, but extract the job title for all jobs on this page. Store the results in a list.

```python
# return a list of tags that contain job titles 
list_of_job_titles_tags = soup.findAll('h2')

list_of_job_titles_tags
```

```python
# get a list of only job titles
list_of_job_titles = []

for x in list_of_job_titles_tags: 
    list_of_job_titles.append(x.text)
    
list_of_job_titles
```

---
1c.  Finally, extract the companies, locations, and posting dates for each job. For example, the first job has a company of "Payne, Roberts and Davis", a location of "Stewartbury, AA", and a posting date of "2021-04-08". Ensure that the text that you extract is clean, meaning no extra spaces or other characters at the beginning or end.

### Companies

```python
# list of company tags
company_tags = soup.findAll('h3')

# list of companies
companies = []

for x in company_tags: 
    companies.append(x.text)
    
companies
```

### Locations

```python
# find all elements where class = 'location' inside the paragraph tags
location_tags = soup.findAll('p', class_ = 'location')
location_tags

# list of locations (stripped)
location = []

for x in location_tags: 
    location.append(x.text.strip())
    
location
```

### Dates

```python
# find all elements where class = 'is-small has-text-grey' inside the paragraph tags
date_tags = soup.findAll('p', class_ = 'is-small has-text-grey')
date_tags

# list of locations (stripped)
date = []

for x in date_tags: 
    date.append(x.text.strip())
    
date
```

---

1d. Take the lists that you have created and combine them into a pandas DataFrame.

```python
jobs_df = pd.DataFrame({'companies' : companies, 'locations' : location, 'dates' : date})
jobs_df
```

# Add a column that contains the url for the "Apply" button. Try this in two ways.

2a. First, use the BeautifulSoup find_all method to extract the urls.

```python
# return all footer tags where the class is 'card-footer'
footers = soup.findAll('footer', class_ = 'card-footer')
footers
```

```python
# return all a tags inside the footer tags
a_tags = []
for x in footers:
    a_tags.append(x.findAll('a', class_ = 'card-footer-item'))
    
a_tags
```

```python
# return only the tags for the apply button
apply_tags = []

for x in a_tags:
    apply_tags.append(x[1])
    
apply_tags
```

```python
# pull out only the urls using a list comprehension
apply_urls = [x.get('href') for x in apply_tags]
apply_urls
```

```python
jobs_df['apply_urls'] = apply_urls
```
